```python
import pyautogui
import time

pyautogui.PAUSE = 1

#1º Step: System's login with a link
pyautogui.hotkey('ctrl', 't')
pyautogui.write('https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema')
pyautogui.press('enter')

#We can put a sleep time to wait the page is full load
time.sleep(5)

#2º Step: Login

## Click in the login space
pyautogui.click(x=919, y=475)

## Login
pyautogui.write('login')

## Password
pyautogui.click(x=940, y=566)
pyautogui.write('password')

## Click to acess
pyautogui.click(x=945, y=662)
time.sleep(3)

#3º Step: Explore the data base

## Click in the data
pyautogui.click(x=494, y=427)

## Click in 3points
pyautogui.click(x=871, y=255)

## Click in download
pyautogui.click(x=978, y=717)

## Save
time.sleep(5)
pyautogui.click(x=739, y=558)

#4ª Step: Calculate the indicators
import pandas as pd

data = pd.read_csv(r"C:\Users\lidiap\Downloads\Compras.csv", sep=";")
display(data)
spent = data['ValorFinal'].sum()
quant = data['Quantidade'].sum()
ticket = spent/quant
print (spent)
print (quant)
print (ticket)

#5º Step: Send a email to the boss 
import pyperclip 

pyautogui.hotkey('ctrl','t')
pyautogui.hotkey("ctrl", "t")
pyautogui.write("https://mail.google.com/mail/u/0/#inbox")
pyautogui.press("enter")

time.sleep(5)

## Click in the write botton
pyautogui.click(x=109, y=249)

## Put the email information
### Remittee
pyautogui.write("lidiaperside@gmail.com")
pyautogui.press("tab") #Choose remittee

### Subject
pyautogui.press("tab") #Subject camp
pyperclip.copy("Sales Report")
pyautogui.hotkey("ctrl", "v")

pyautogui.press("tab") #Email camp

text = f"""
Dear,
Follow the purchase report

Total Spent: R${spent:,.2f}
Quantity of Products: {quant:,}
Average Price: R${ticket:,.2f}

If you have any questions, do not hesitate to ask.
Att., Lidia Perside
"""

pyperclip.copy(text)
pyautogui.hotkey('ctrl', 'v')

### Send the email
pyautogui.hotkey('ctrl','enter')
```
